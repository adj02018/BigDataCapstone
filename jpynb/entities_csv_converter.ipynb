```python
#utf-8형식으로 해야됩니다.
filenames = 'professor' # 파일명입니다. 확장자명은 기본값은 csv입니다.

filename = filenames + ".csv"
converter = filenames + "_cvt.csv"
f = open(filename, 'r', encoding='UTF-8') # 파일을 읽기용으로 사용합니다.

csvarray = [] # 새로이 큰따옴표를 넣게할 문자열형식의 배열입니다.

dswitch = 0 # csv문서가 이미 따옴표가 있는지 확인여부하여 작동할지 결정합니다. 
# 0 따옴표가 없어서 작동해야됨 || 1 따옴표가 있어서 작동하지 말아야됨

line = f.readline()

if line[-2] == '"':
    dswitch = 1

if dswitch == 0:
    while True:
        line = f.readline()
        if not line: # 읽지 못하면 반복문 빠져 나옵니다.
            break
        line = line.replace(',','","')#print('"'+line.rstrip("\n")+'"\n')
        csvarray.append('"'+line.rstrip("\n")+'"\n')
    f.close()
    f = open(converter, 'w', encoding='UTF-8') # 파일을 쓰기용으로 사용합니다.
    for cell in csvarray:
        print(cell)
        f.write(cell)
else:
    print("이미 큰따옴표가 기입된 문서입니다.")
    
f.close()

```
